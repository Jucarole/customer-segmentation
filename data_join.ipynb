```python
import pandas as pd
```

```python
data = pd.read_csv('./data/new_data.csv')
clusters = pd.read_csv('./data/customers_clusters.csv')
```

```python
print('data shape: ', data.shape)
print('clusters shape: ', clusters.shape)
```

```python
data_clusters = data.join(clusters.set_index('CustomerID')[['Cluster']], on = 'CustomerID')
```

```python
clusters['Cluster'].value_counts()
```

```python
data_clusters.groupby('CustomerID')['Cluster'].min().value_counts()
```

```python
data_clusters['Cluster'].isna().value_counts()
```

```python
data_clusters['Product'].value_counts()
```

```python
clusters.plot.scatter(x='PurchaseCount', y='PurchaseSum', c='Cluster', colormap='viridis')
```

```python
data_clusters.plot.scatter(x='Quantity', y='Total', c='Cluster', colormap='viridis')
```

```python
data_clusters.plot.scatter(x='Quantity', y='Total', c='Product', colormap='plasma')
```

```python
data_clusters.to_csv('./data/data_clusters.csv', index=False)
```
